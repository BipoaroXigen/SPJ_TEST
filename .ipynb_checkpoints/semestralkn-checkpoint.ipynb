```julia
using EO
using Plots
```

```julia
#= basis_functions = [sin, cos, +, -, *, /, x->x^2, (x,y)->x^xy]
basis_variables = [1, 0] =#
basis_functions = [+, -, *, /]
basis_variables = Vector{Any}([1, 0, :var1])

true_f = x -> x^2

x = Float64.(collect(1:100));
y = true_f.(x);

plot(x, y, label="true_f")
```

```julia
pop_size = 30
dimension = 1

objective_function  = enclose_arguments(EO.f_function_diff, x, y)
initialization      = enclose_noargs(EO.expression_initialization, pop_size, objective_function, basis_functions, basis_variables)
selection           = enclose_arguments(EO.s_tournament, 10, 3)
crossover           = enclose_arguments(EO.cr_subtree, pop_size, basis_functions, basis_variables)
mutation            = enclose_arguments(EO.subtree_mutation!, basis_functions, basis_variables)
replacement         = EO.enclose_replacement(r_keep_best_n, pop_size)
termination         = enclose_argument(iteration_termination, pop_size*10)

solution = solvink_hart(objective_function, initialization, selection, crossover, mutation, replacement, termination)

f = eval(:(x -> $(solution.top_coords)));
yy = @eval $f.($x);
```

```julia
plot(x, yy, label="regressed_f")
```

```julia
get_ast_len(e::Expr) = sum(get_ast_len.(e.args))
get_ast_len(e::Any) = 1
get_symbols(e::Expr) = [get_symbols(e_i) for e_i in e.args]
get_symbols(e::Any) = e

function Base.getindex(e::Expr, i::Int)
    dump(e)
    symbols = get_symbols(e)
    display(symbols)
    symbols =  reduce(vcat, symbols)
    display(symbols)
    return symbols[i]
end

function Base.setindex!(e::Expr, value::Union{Expr, Symbol}, i::Int)
    dump(e)
    symbols = get_symbols(e)
    display(symbols)
    # Flatten the nested array of symbols
    symbols = reduce(vcat, symbols)
    display(symbols)
    # Keep track of how many elements we've skipped
    skipped = 0
    for arg in e.args[1:i-1]
        if arg isa Expr
            skipped += length(arg.args) - 1
        end
    end
    i -= skipped
    dif_i = 0


    if symbols[i] isa Symbol
        e.args[i] = value
    else
        i = i - dif_i
        e.args[i] = Expr(:call, value, e.args[i])
    end
    return e
end

```

```julia
function getindex(ex::Expr, idx::Int)
    if idx == 0
        ex
    else
        csize = cumsum(map(nodes, ex.args[2:end]))
        nidx = findfirst(i->i>=idx, csize)
        nex = ex.args[nidx+1]
        !isa(nex, Expr) ? nex : nex[csize[nidx]-idx]
    end
end

function setindex!(ex::Expr, subex, idx::Int)
    csize = cumsum(map(nodes, ex.args[2:end]))
    nidx = findfirst(i->i>=idx, csize)
    nex = ex.args[nidx+1]
    if !isa(nex, Expr) || csize[nidx]-idx == 0
        ex.args[nidx+1] = subex
    else
        nex[csize[nidx]-idx] = subex
    end
end
```
