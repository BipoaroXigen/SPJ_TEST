```julia
using EO

#TSP_graph, v = EO.parse_TSP_problem("tsp_problems/a280");
TSP_graph, v = EO.parse_TSP_problem("ALL_tsp/berlin52");
dimension = size(TSP_graph, 1)
#TSP_graph, v = EO.parse_TSP_problem("ALL_tsp/a280");
objective_function  = enclose_arguments(EO.f_dist_sum, TSP_graph)
initialization      = enclose_noargs(EO.TSP_NN_initialization, dimension, 2, objective_function, TSP_graph)
crossover           = EO.cr_edge_recombination

pop = initialization()

EO.plot_graph_sequence(pop.population[2].sequence, v)
```
